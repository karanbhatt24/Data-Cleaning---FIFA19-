```python
# Step 1 - Import required Libraries
import pandas as pd
import numpy as np
import matplotlib as plt
```

```python
# Step 2 - Import the dataset
dataset = pd.read_csv(r'C:\Users\sunee_000\Desktop\FIFA\fifa.csv')
#used 'r' before the file path to convert normal string to raw string
```

```python
# Step 4 - Look for duplicacy
# Check for duplicate rows
DuplicateRows = dataset.duplicated()
sum(DuplicateRows)
```

```python
# Step 5 - Dropping the unnecessary variables
# Print the list of column names 
print(list(dataset.columns))
```

```python
# Dropped the unnecessary variables mentioned below
dataset1 = dataset.drop(['Unnamed: 0','ID','Photo','Flag','Club Logo','Real Face','Jersey Number','Loaned From'], axis=1)
# axis = 1 refers to dropping columns
```

```python
dataset1.shape # Reduced the number of columns from 89 to 81 
```

```python
# Step 6 - Dealing with missing values
dataset1.isnull().sum()
```

```python
# Dropping further variables that does not contribute towards sense making of the data
dataset2 = dataset1.drop(['Special','Work Rate', 'Body Type','Joined','Contract Valid Until','Height', 'Weight','LS', 'ST', 
                          'RS', 'LW', 'LF', 'CF', 'RF', 'RW', 'LAM','CAM', 'RAM', 'LM', 'LCM', 'CM', 'RCM', 'RM', 'LWB',
                          'LDM', 'CDM', 'RDM', 'RWB', 'LB', 'LCB', 'CB', 'RCB', 'RB','Release Clause'], axis=1)
```

```python
dataset2.shape # Reduced the variables to 49 from 81 through domain knowledge
```

```python
dataset2.isnull().sum()
```

```python
dataset2.head()
```

```python
# Remove the rows with missing values as missing records accounts for very minor portion of whole dataset records i.e 18207
dataset3 = dataset2.dropna(axis=0) #axis = 0 is for dropping the rows in the dataset
```

```python
dataset3.shape # Dropped the missing records and reduced the dataset records from 18207 to 17918 i.e dataloss 0.0158% 
```

```python
dataset3.isnull().sum() # Checking for missing values 
```

```python
dataset3.head()
```

```python
# Step 7 - Checking for data types of each variable and converting the datatypes according to the requirements
dataset4=pd.DataFrame(dataset3) # Transferred data treated with missing values to another dataframe to maintain data hygiene
```

```python
dataset4.shape
```

```python
dataset4.head()
```

```python
dataset4.dtypes
```

```python
dataset4['Value'] = dataset4['Value'].str.replace('\u20AC','')
# Removing euro symbol from wage and value, ASCII value for euro symbol is '\u20AC
dataset4['Wage'] = dataset4['Wage'].str.replace('\u20AC','')
```

```python
dataset4.head()
```

```python
dataset4['Value'] = dataset4['Value'].replace({'K': '*1e3', 'M': '*1e6'}, regex=True).map(pd.eval).astype(int)
# Replacing the records for Value with K by multipying 1000 instead and same w.r.t M by multiplying 1000000
```

```python
dataset4['Wage'] = dataset4['Wage'].replace({'K': '*1e3', 'M': '*1e6'}, regex=True).map(pd.eval).astype(int)
# Replacing the record for Wage with K by multipying 1000 instead and same w.r.t M by multiplying 1000000
```
